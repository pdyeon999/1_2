```python
# 기본 실습
#1. DataFrame 객체 생성
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age':[25, 30, 35, 40, 28],
    'City':['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary':[70000, 80000, 120000, 90000, 75000]
})
df
```

```python
#2. 열과 행 선택

print(df.iloc[:, [0, 1]])
print(df[df['Age']>=30])
```

```python
#3 행과 열 추가(1)

df['Country']='USA'
df
```

```python
#3 행과 열 추가(2)

df.loc[len(df)] = ['Frank', 33, 'Seattle', 85000, 'USA']
df
```

```python
#4 데이터 삭제(1)
df.drop('Salary', axis=1, inplace=True)
df
```

```python
#4 데이터 삭제(2)
df.drop(df.index[df['Name']=='David'], axis=0, inplace=True)
df
```

```python
#5 데이터 수정(1)
df.loc[df['Age']>=35, 'Salary'] *= 1.05
df
```

```python
# 데이터 수정(2)
df.loc[df['Name']=='Alice', 'City'] = 'Boston'
df
```

```python
# 데이터 타입 변환
df['Age'] = df['Age'].astype(float)
df['Salary'] = df['Salary'].astype(float)
df['Age'].dtype, df['Salary'].dtype
```
