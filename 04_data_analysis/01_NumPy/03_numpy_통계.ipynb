```python
import numpy as np
```

```python
height = np.array([150, 155, 160, 164, 165, 170, 170, 173, 175, 175, 175, 180, 180, 185, 189])
```

```python
values, count = np.unique(height, return_counts=True)
print(count)
print(np.max(count))
print(values[count == np.max(count)])
```

```python
print(np.var(height))
print(np.std(height))
print(np.sqrt(height.var()))
```

```python
print(np.max(height)-np.min(height))
```

```python
# 확률분포
import matplotlib.pyplot as plt
```

```python
arr = np.random.randn(1000)
plt.hist(arr, bins=1000)
plt.show()
print(arr.mean(), arr.std())
```

```python
arr = np.random.normal(10, 5, (10, 10))
plt.hist(arr, bins=30)
plt.show()
print(arr)
print(arr.mean(), arr.std())
for i in range(10):
    print(arr[i].mean(), arr[i].std())
```

```python
arr = np.random.uniform(1, 3, (5, 10))

plt.hist(arr, bins=10)
plt.show()
```

```python
np.random.seed(0)
arr = np.random.randint(1, 100, (100, 20))
print(arr)
plt.hist(arr, bins=100)
plt.show()
```
