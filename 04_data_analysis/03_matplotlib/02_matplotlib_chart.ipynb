### matplotlib 차트 유형

1. histogram : 연속형값의 도수(빈도수/개수) 파악
2. bar plot : 범주형값의 개수, 또는 다른 연속값을 파악
3. scatter plot : 산포도(산점도) 두 연속값들의 분포 파악
4. line plot : 시계열데이터(연속값) 변화 파악
5. box plot : 연속형 값의 사분위, 이상치 파악
6. pie chart : 전체 범위 중의 비율

```python
# Histogram: 연속형 값을 일정 구간을 나눠 구간별 데이터포인트의 빈도를 시각화
# -- x축: 연속형 데이터
# -- y축: 구간별 데이터포인트의 개수 (합)

scores = [100, 95, 97, 98, 99, 90, 80, 81, 84, 88, 70, 78, 79, 70, 65, 66, 61, 64, 100, 89]
bins = [60, 70, 80, 90, 100]

plt.hist(scores, bins=bins, edgecolor = 'pink', rwidth=0.5)
#bins: 나뉠 구간의 개수 (기본값: 10)
#edgecolor: 히스토그램 막대의 경계선 색상
#rwidth: 0~1 사이값. 상대적 너비 지정

plt.xlabel('점수')
plt.ylabel('인원수')
plt.title('점수별 인원도수표')
plt.show()
```

```python
# 누적 히스토그램

plt.hist(scores, edgecolor = 'pink', rwidth=0.5, cumulative=True)
# cumulative: 누적값으로 표시

plt.show()
```

```python
# bar plot: 막대 차트
# 범주형 데이터와 연속형 데이터(막대의 길이)의 관계를 시각화
# -- x축 : 범주형 데이터 카테고리
# -- y축 : 연속형 데이터 (총합, 평균)
```

```python
# 월별 매출 데이터
df = pd.DataFrame({
    'months': ['1월', '2월', '3월', '4월', '5월', '6월'],
    'sales': [1300, 1600, 1700, 1650, 1900, 2300]
})

plt.bar(df['months'], df['sales'], zorder=10)
#zorder : z축 깊이 설정. 값이 클수록 앞으로 나옴.

for i, val in enumerate(df['sales']):
    plt.text(i, val+100, str(val), ha='center')
# 각 막대마다 라벨링. x축 순서, y축 위치, 나타낼 값(텍스트), 텍스트 정렬 위치
# str() 없으면 숫자형으로 잘 출력됨

plt.xlabel('월')
plt.ylabel('매출(단위: 천 만원)')
plt.grid(zorder=1)

plt.show()
```

```python
# 수평 막대 그래프
# -- x축: 연속형
# -- y축: 범주형

plt.barh(df['months'], df['sales'])

plt.ylabel('월')
plt.xlabel('매출(단위: 천 만원)')

plt.show()
```

```python
# 주가 추이
df = pd.DataFrame({
    'Date' : pd.date_range(start='2025-01-01', periods=14, freq='D'),
    # 1월 1일부터 14일만큼을 연속된 날짜로 Day 단위로 올림
    'APPL': np.random.randint(100, 150, 14),
    'MSFT': np.random.randint(100, 150, 14)
})
df
```

```python
plt.plot(df['Date'], df['APPL'], label='APPL')
plt.plot(df['Date'], df['MSFT'], label='MSFT')

plt.xlabel('날짜')
plt.ylabel('주가')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# 해당 데이터프레임에서 plot 그릴 것

df.plot(
    x='Date',
    y=['APPL', 'MSFT'],
    color=['black', 'yellow'],
    label=['apple', 'microsoft'],
    # y축 값이 두 개 이므로 color, label도 모두 두 개씩
    subplots=True
    # 두 개의 그래프가 별도로 나옴(subplot형태)
)

plt.xlabel('날짜')
plt.ylabel('주가')
plt.show()

#legend() 안 해도 label 지정했으면 자동으로 띄움
```

```python
# pychart
# 전체에서 각 부분이 기여하는 비율을 시각화

df = pd.DataFrame({
    'label': ['애플', '삼성', '샤오미', 'LG'],
    'market_share': [33, 32, 10, 25], 
    'sales': [1345, 234, 545, 454],
    'explode': [0, 0.1, 0, 0]
})

plt.pie(
    df['market_share'],
    # 어떤 값의 비율을 나타낼지
    labels=df['label'],
    # 라벨로 띄울 것 지정
    autopct='%.1f%%',
    # 비율값을 소숫점 아래 한 자리까지 나타냄
    startangle=90, #원판 돌림
    explode=df['explode'], #중심으로부터 얼마나 떨어질지
    colors=['skyblue', 'lightgreen', 'lightcoral', 'gold']
)

plt.title('기업별 시장 점유율')
plt.show()
```

```python
plt.pie(
    df['sales'],
    labels=df['label'],
    autopct='%.1f%%',
    startangle=90,
    explode=[.05, .05, .05, .05],
    wedgeprops={'width': .6}
    # 가운데 뚫음
)
plt.text(0, 0, '기업별 판매 비율', ha='center', va='center')
# text를 중앙에 배치

plt.show()
```

```python
# Box plot
# 연속형 데이터의 분포를 사분위수로 시각화
# 이상치 outlier를 도출하는 데에 효과적
# -- x축: 이산값(범주형) 데이터
# -- y축: 연속형 데이터
  
# - 표현되는 값들
#  - max, Q3, Q2, Q1, min(최대값, 3분위수(75%), 2분위수(50%, 중위값), 1분위수(25%), 최소값)
#  - IQR (Inter-quartile Range) : 사분범위
```

```python
# 반별 시험 성적 데이터 (2차원 배열)
class_scores = [
    [85, 88, 90, 92, 93, 87, 95, 91, 84, 86],  # Class A
    [78, 82, 85, 80, 79, 84, 81, 83, 77, 76],  # Class B
    [92, 95, 96, 94, 90, 97, 93, 91, 98, 99],  # Class C
    [70, 72, 68, 75, 74, 69, 73, 71, 66, 67]   # Class D
]

plt.boxplot(class_scores)


plt.xlabel('반')
plt.ylabel('점수')
plt.show()
# 각 리스트에 대해 값의 범위를 표시.
# 직선으로 최대,최소 표시. 박스로 Q1, Q3 표시. 중앙선으로 중앙값 표시
```

```python
data = class_scores[0]

print('최소값:', np.min(data))
print('사분위수(Q1):', np.percentile(data, 25))
# percentile: data 데이터 중 25% 값을 가져옴. 0.25가 아닌 25(int형으로)
print('사분위수(Q2):', np.percentile(data, 50), np.median(data))
print('사분위수(Q3):', np.percentile(data, 75))
print('최대값:', np.max(data))
```

```python
weird_data = [
    [100, 120, 130, 110, 115, 105, 125, 200],   # 이상치 200   
    [150, 160, 170, 155, 165, 158],
    [80, 85, 90, 75, 92, 87, 300],   # 이상치 300
    [200, 220, 210, 205, 215, 230, 50]  # 이상치 50    
]

# whis 기본값 : 1.5
# [참고] 수염의 길이 계산 공식 : [Q1 - whis * IQR, Q3 + whis * IQR]
# [참고] IQR = Q3 - Q1
plt.boxplot(weird_data, whis=5)
# whis 기본값은 1.5. 이상치를 boxplot이 판단하는 데 가중치임.
# 5로 크게 줘서 상품1의 데이터 200은 이상치(동그라미)로 표현X

plt.xticks([1, 2, 3, 4], ['상품1', '상품2', '상품3', '상품4'])
plt.xlabel('상품명')
plt.ylabel('싯가')
plt.show()

#자동으로 수염 길이 계산해 벗어나는 이상치는 동그라미로 표시
```
