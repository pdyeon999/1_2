```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
```

```python
font_path = 'C:/Windows/Fonts/gulim.ttc'
font = fm.FontProperties(fname=font_path).get_name()
matplotlib.rc('font', family=font) #폰트 설정 변경
```

```python
x=[1, 2, 3]
y=[4, 5, 6]
plt.plot(x, y)
plt.show()
```

```python
### Figure, Axes

fig = plt.figure(figsize=(6,4), facecolor='yellow')
ax = plt.gca()      # .gca()로 axes객체 반환
ax.plot(x, y)

print(type(fig), type(ax))

plt.title('Hello Matplotlib')
plt.xlabel('X')     #x축의 label 지정
plt.ylabel('Y')     #y축의 label 지정
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 6))
plt.show()
print(fig)
print(ax)
#nrows: 그려질 그래프 영역의 행 수
#ncols: 그려질 그래프 영역의 열 수
```

```python
x=np.arange(1, 11)
y=x*2
plt.plot(x, y)
plt.show()
```

```python
x=pd.Series(x)
y=pd.Series(y)
plt.plot(x, y)
plt.show()
```

```python
x = pd.DataFrame({'x':x})
y=pd.DataFrame({'y':y})
plt.plot(x, y)
plt.show()
```

```python
# plot 메서드의 매개변수

plt.figure(figsize=(6, 4))
plt.plot(
    x, y,
    color='black',
    marker='o',
    markersize=4,
    linestyle='--',
    linewidth=2
)
plt.show()
```

```python
plt.figure(figsize=(3, 4))
x=np.arange(1, 101)
y=np.linspace(2, 200, 100)

plt.xticks(ticks=np.arange(0, 101, 5), rotation=45)
plt.yticks(rotation=45)
plt.plot(x, y)
plt.show()
```

```python
plt.figure(figsize=(3, 2))

x=np.arange(1, 101)
y=np.linspace(2, 200, 100)

plt.xlim(1, 120)
plt.ylim(2, 100)

plt.plot(x, y)
plt.show()
```

```python
#legend (범례)
# upper|rower , left|right ,, center
plt.figure(figsize=(3, 2))
x=np.arange(1, 101)
y=np.linspace(2, 200, 100)
y2=np.linspace(3, 300, 100)

plt.plot(x, y, label='시간')
plt.plot(x, y2, label='공간')
plt.legend(loc='lower right')
plt.show()
```

```python
# 그래프 여러 개 그리기

plt.figure(figsize=(3, 2))

x=np.arange(1, 11)
y=x*3
y2=x*2

plt.plot(x, y, label='선', color='red', marker='o')
plt.bar(x, y2, label='바', color='green')

plt.legend()
plt.grid()

plt.show()
```

```python
# subplots 이용 -> 여러 그래프를 각각 그림
fig, ax = plt.subplots(1, 2, figsize=(18, 5))

x1 = np.arange(1, 10)
y1 = x1 * 2
x2 = np.arange(11, 20)
y2 = x2 * 2

ax[0].plot(x1, y1, label='선', color='red')
ax[0].set_xlabel('x1')
ax[0].set_ylabel('y1')
ax[0].set_title('빨간선')
ax[0].legend()
ax[0].grid(True)

ax[1].bar(x2, y2, label='바', color='yellowgreen')
ax[1].set_xlabel('x2')
ax[1].set_ylabel('y2')
ax[1].set_title('노랑연두선')
ax[1].legend()
ax[1].grid(True)


plt.show()
```
