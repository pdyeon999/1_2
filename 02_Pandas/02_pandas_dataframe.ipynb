```python
# 행 추가 및 삭제

students = [
    {'name':'호랑이', 'midterm':95, 'final':85},
    {'name':'늑대', 'midterm':93, 'final':90},
    {'name':'양', 'midterm':100, 'final':10}
]
df=pd.DataFrame(students)
df
```

```python
#1. loc 이용
df.loc[len(df)] = ['부엉이', 88, 72]
df
```

```python
#2. concat 이용

df = pd.concat([df, pd.DataFrame([{'name':'곰', 'midterm':99, 'final':24}])], ignore_index = True)
df
```

```python
# 행 삭제
#1 인덱스 이용
df.drop(df.index[[0]], inplace=True)
df
```

```python
# 컬럼 추가
df = pd.DataFrame({
    '이름':['토끼', '판다', '코알라'],
    '위치':['독산', '종로', '하남'],
    '성별':['F', 'M', 'F']
})

#1 기본값 이용
df['취미'] = ['명상', '줄넘기', '나무']
df
```

```python
# ndarray이용 np.where()
df['성별(한글)'] = np.where(df['성별']=='M', '남성', '여성')
df
```

```python
df = pd.DataFrame({
    '이름':['토끼', '판다', '코알라'],
    '위치':['독산', '종로', '하남'],
    '성별':['F', 'M', 'F'],
    '키':[179, 165, 157],
    '체중':[50.1, 48.2, 51.3]
})
df
```

```python
# BMI 공식: kg / m^2
df['BMI'] = df['체중'] / (df['키']/100)**2
df
```

```python
# apply()

def get_type(value):
    return '저체중' if value<18 else '표준'

df['BMI type'] = df['BMI'].apply(get_type)
df
# 함수 전달
```

```python
# 컬럼 삭제
# drop 이용

df.drop('BMI type', axis = 1, inplace=True)
df
```
