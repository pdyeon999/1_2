```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['토끼', '판다', '코알라', '양']
})

df2 = pd.DataFrame({
    'id':[1,2,3,5],
    'score':[100, 95, 88, 71]
})
```

```python
pd.merge(df1, df2, on='id')
```

```python
pd.merge(df1, df2, on='id', how='left')
```

```python
pd.merge(df1, df2, on='id', how='right')
```

```python
pd.merge(df1, df2, on='id', how='outer')
```

```python
df1 = pd.DataFrame({
    'n_id':[1,2,3,4],
    'name':['토끼', '판다', '코알라', '양']
})

df2 = pd.DataFrame({
    's_id':[1,2,3,5],
    'score':[100, 95, 88, 71]
})
```

```python
pd.merge(df1, df2, left_on='n_id', right_on='s_id', how='left')
```

```python
# join(특정 인덱스를 기준으로 DataFrame 병합)

df3 = pd.DataFrame({
    'age':[33, 45, 22, 20],
    'city':['서울', '수원', '인천', '서울']
}, index=['이석훈', '김용준', '김진호', '이주원'])
df4 = pd.DataFrame({
    'pet':['고양이', '강아지', '햄스터', '돌']
}, index=['이석훈', '김용준', '김진호', '주원'])

print(df3)
df4
```

```python
df3.join(df4, how='inner')
```

```python
df3.join(df4, how='left')
```

```python
df3.join(df4, how='right')
```

```python
df3.join(df4, how='outer')
```

```python
## CONCAT

df5 = pd.DataFrame({
    'name': ['구준표', '윤지후', '소이정', '프린스송'],
    'age': [18, 18, 18, 18]
})
df6 = pd.DataFrame({
    'name': ['금잔디'],
    'age': [17]
})
```

```python
pd.concat([df5, df6], axis=0).reset_index(drop=True)
```

```python
# MELT

df = pd.DataFrame({
    'name': ['홍길동', '신사임당', '이순신'],
    'math': [80, 99, 100],
    'english': [34, 58, 48],
    'science': [99, 55, 77],
})
```

```python
pd.melt(
    df,
    id_vars=['name'],    #고정(유지)될 칼럼
    value_vars=['math', 'english', 'science'],  #합쳐질 칼럼 지정
    var_name='subject',  #칼럼명이 합쳐질 컬럼 이름 (기본값 variable)
    value_name='score'   #칼럼값이 합쳐질 칼럼의 이름(기본값 value)
)
```
